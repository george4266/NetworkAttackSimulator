````python
import plotly.express as px
import pandas as pd
````

````python
sheetID = "1kXe3UEXWdoJdQIQlDxROsovTpaiAjON3-j7X6QtM_ls"
df = pd.read_csv("Q-l agent_out_2.csv")
df = pd.DataFrame(df)

df.drop(df.columns[0], axis=1, inplace=True) #first columns is duplicate created when creating the original DataFrame
print(df)

#https://docs.google.com/spreadsheets/d/1kXe3UEXWdoJdQIQlDxROsovTpaiAjON3-j7X6QtM_ls/edit?usp=sharing
````

````python

#3d scatter plot
fig = px.scatter_3d(df, x="reward", y="diff_in_reward", z = "steps", color="diff_in_reward", title= "3D Scatter showing Difference in Rewards between Rounds", symbol="diff_in_reward", opacity=0.7, template="simple_white")

fig.show()
````

### The Above 3D Chart...

Has shown that there are differences in rewards between steps. The majority of difference in rewards between steps is greater than -2 (most are -1). 

````python

#3d scatter plot
fig2 = px.scatter_3d(df, x="reward", y="diff_in_reward", z = "steps", color="Actions", title= "3D Scatter showing Difference in Rewards between Rounds", symbol="Actions", opacity=0.5, template="simple_white", hover_data={
    
    "reward" : False,
    "diff_in_reward" : False,
    "steps": True,
    "Actions" : False})


fig2.show()
````

## Below is a Correlation Heatmap...

I am in the process of looking for correlation in the data between rewards, difference in rewards, and actions. This uses the .corr() method instead of my previous "looking method" :P



````python
import numpy as np
import plotly.graph_objects as go


df_corr = df.corr()

print(df_corr)

````

````python
fig = go.Figure()
fig.add_trace(
    go.Heatmap(
        x= df_corr.columns,
        y=df_corr.index,
        z = np.array(df_corr),
    )
)
fig.show()
````
